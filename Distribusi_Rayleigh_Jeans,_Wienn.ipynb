```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
h = 6.626E-34
k = 1.3807E-23
c = 3E8
pi = np.pi
e = np.e

Ta = 5800
Tb = 10000
Tc = 30000
```

```python
def RJf(f,pi,k,T,c):
  return (8*pi*(f**2)*k*T)/c**3

def RJl(l,pi,k,T,c):
  return (8*pi*k*T)/(l**4)

def Wf(f,pi,h,k,T,c):
  return ((8*pi*h*(f**3))/(c**3))*e**(-h*f/(k*T))

def Wl(l,pi,h,k,T,c):
  return ((8*pi*h*c)/(l**5))*e**(-h*c/(l*k*T))

def Pf(l,pi,h,k,T,c):
  return ((8*pi*h*(f**3))/(c**3))*(1/((e**(h*f/(k*T)))-1))

def Pl(l,pi,h,k,T,c):
  return ((8*pi*h*c)/(l**5))*(1/((e**(h*c/(l*k*T)))-1))
```

```python
lmax = 3E-6
lmin = 1E-15
l = np.linspace(lmin,lmax,1000)

fmax = 5E15
fmin = 1
f = np.linspace(fmin,fmax,1000)

lx = l * 1E9
fx = f * 1E-12

RJfa, RJfb, RJfc = RJf(f,pi,k,Ta,c), RJf(f,pi,k,Tb,c), RJf(f,pi,k,Tc,c)
RJla, RJlb, RJlc = RJl(l,pi,k,Ta,c), RJl(l,pi,k,Tb,c), RJl(l,pi,k,Tc,c)
Wfa, Wfb, Wfc = Wf(f,pi,h,k,Ta,c), Wf(f,pi,h,k,Tb,c), Wf(f,pi,h,k,Tc,c)
Wla, Wlb, Wlc = Wl(l,pi,h,k,Ta,c), Wl(l,pi,h,k,Tb,c), Wl(l,pi,h,k,Tc,c)
Pfa, Pfb, Pfc = Pf(l,pi,h,k,Ta,c), Pf(l,pi,h,k,Tb,c), Pf(l,pi,h,k,Tc,c)
Pla, Plb, Plc = Pl(l,pi,h,k,Ta,c), Pl(l,pi,h,k,Tb,c), Pl(l,pi,h,k,Tc,c)
```

#Distribusi 5800 K


```python
plt.figure(figsize=(10, 6))
plt.plot(lx, RJla, label=f'Rayleigh-Jeans Distribution', color='Red')
plt.plot(lx, Wla, label=f'Wienn Distribution', color='Green')
plt.plot(lx, Pla, label=f'Planck Distribution', color='Blue')

plt.xlabel('Wavelength (nm)')
plt.ylabel('Intensity')
plt.title(f'Wavelength Distribution at {Ta} K')
plt.legend()
plt.grid(True)
plt.xlim(50, 2000)
plt.ylim(0, np.max(Pla)*1.1)

plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.plot(fx, RJfa, label=f'Rayleigh-Jeans Distribution', color='Red')
plt.plot(fx, Wfa, label=f'Wienn Distribution', color='Green')
plt.plot(fx, Pfa, label=f'Planck Distribution', color='Blue')

plt.xlabel('Frequency (10^12 Hz)')
plt.ylabel('Intensity')
plt.title(f'Frequency Distribution at {Ta} K')
plt.legend()
plt.grid(True)
plt.xlim(0, 2000)
plt.ylim(0, np.max(Pfa)*1.1)

plt.show()
```
